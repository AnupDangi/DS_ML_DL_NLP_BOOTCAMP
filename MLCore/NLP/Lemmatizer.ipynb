## Lemmatization
Lemmatization is an important text pre-processing technique in Natural Language Processing (NLP) that reduces words to their base form known as a "lemma." For example, the lemma of "running" is "run" and "better" becomes "good."



```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('going')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
"""
POS- NOUN -V
verb-v 
adjactive -a
adverb -r
"""
```

```python
lemmatizer.lemmatize("going",pos='v')
```

```python
lemmatizer.lemmatize("going",pos='a')
```

```python
lemmatizer.lemmatize("going",pos='')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","going","go"]
```

```python
for word in words:
    print(word +"---------->"+lemmatizer.lemmatize(word,pos='v'))
```
