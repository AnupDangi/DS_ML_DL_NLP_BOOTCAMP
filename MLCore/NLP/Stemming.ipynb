Stemming

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma.
Stemming is important in natural lanugage understanding (NLU) and natural langauge processing.

```python
## Classification Problem 
## Comments of products in a positive review or negative review
## Reviews ---> eating ,eat,eaten [ going,gone,goes]
```

RegexpStemmer class


```python
words=["eating","eats","eaten","writing","writes","programming","programs","going","go"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word +"-------->"+ stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
## Snowball Stemmer
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+ "---> "+snowballstemmer.stem(word))
```

```python
stemming.stem('fairly'),stemming.stem('sportingly')
```

```python
snowballstemmer.stem('fairly'),snowballstemmer.stem('sportingly')
```

```python
snowballstemmer.stem('going')
```
