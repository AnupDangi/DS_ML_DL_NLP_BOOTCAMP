```python
## Let's load the dataset and understand its structure
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

data=pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
data.head()
```

```python
data.info()
```

## Data Cleaning and Preprocessing



```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

```python
## we create a column region 
df=data
df.loc[:122,'region'] = 0
df.loc[122:,'region']=1

```

```python
df.head()
```

```python
df['region'] = df['region'].astype('int')
```

```python
df.info()
```

```python
## remove the null values 
df.dropna(inplace=True)
```

```python
df.head(125)
```

```python
df.isna().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(123).reset_index(drop=True)
```

```python
df.head(125)
```

```python
df.columns
```

```python
## fix space in columns names 

df.columns=df.columns.str.replace(' ','')
```

```python
## fix the whitespace in columns names
df.columns
```

```python
## Change the required columns as integer data type 
```

```python
df[['month','day','year','region','Temperature','RH','Ws']]=df[['month','day','year','region','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the other columns to float data type 

objects=[features for features in df.columns if df[features].dtype==object]

```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.info()
```

```python
## save as cleaned dataset 

df.to_csv("Algerian_forest_fires_dataset_CLEANED.csv",index=False)
```
