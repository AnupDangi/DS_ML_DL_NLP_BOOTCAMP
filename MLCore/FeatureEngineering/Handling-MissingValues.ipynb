Missing Completely at Random,MCAR:

Missing completely at random (MCAR) is a type of missing data mechanism in which the probability of a vlaue being missing is unrelated to both teh boseved data and the missing data. In other words, if the data is MCAR, the missing values are randomly distributed throughout the dataset,and there is no systematic reason for why they are missing.

for example, in a survey about the prevalance of a certain disease, the missing data might be MCAR if the survery participants with missing values for certain questions were selected randomly and their missing response are not related to their disease status or any other variables measured in the survey.


2. Missing at Random MAR:
it is a type of missing data mechanism in which the probability of a value being missing depends only on the observed data, but not on the missing data itself.In other words, if the data is MAR, the missing values are systematically related to the observed data,but not to the missing data.Here are a few examples of missing at random.

3. Missing data not at random (MNAR)

it is type of missing data mechanism where the probability of missing values depends on the value of the missing data itself. In other words,if the data is MNAR,the missingness is not random and is dependent on unobserved or unmeasured factors that are associated with the missing value.
