```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## Create a line plot
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")
plt.show()
```

```python
# x=[1,2,3,4,5]
# y=[1,4,9,16,25]


## create customized line plot
plt.plot(x, y,color='red', linestyle='--')
```

```python
# plt.plot(x,y,color='red',linestyle='--',marker='o')
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize='9')
plt.grid(True)
```

```python
## Multiple Plots

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1) ## how many rows, columns,height
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,y1,color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y1,color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y1,color='yellow')
plt.title("Plot 3")
```

```python
## Bar Plot 

categories=['A','B','C','D','E']
values=[5,7,3,8,6]

## Create a bar plot
plt.bar(categories,values,width=0.8,color='purple')
plt.xlabel("Categories")
plt.ylabel("Values")
```

```python
## Histogram
## It is used to represent the distribution of a dataset. They divide the data into bins and counts the number of data points in each bins.
```

```python
data=[1,2,3,2,3,3,4,4,4,4,5,5,6,6,6,7,8,8,8,8,9,10]

## Create a histogram
plt.hist(data,bins=5)
```

```python
## Create a scatter plot
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.scatter(x,y,color='red',marker='x')
```

```python
## pipe chart
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
## Sales Data Visualization
import pandas as pd
sales_data=pd.read_csv("sales_data.csv")

sales_data.head(5)
```

```python
## Let's group by product category

product_category=sales_data.groupby('Product Category')['Total Revenue'].sum()
product_category.plot(kind='bar',color='teal')
```

```python
## plot sales trend over time

sales_trend=sales_data.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
